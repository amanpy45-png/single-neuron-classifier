```python
import math
def Sigmoid(x):
    return 1 / (1 + math.exp(-x));
```

```python
# Sigmoid(-1.4)
```

```python
def neuron_forward(x):
    z=np.dot(w,x)+b
    y_hat=Sigmoid(z)
    return y_hat
```

```python
new_student = np.array([7.5, 9])
pred = neuron_forward(new_student)
pred_class = 1 if pred >= 0.5 else 0

print(f"Predicted probability of passing: {pred:.2f}")
print(f"Predicted class: {'Pass' if pred_class==1 else 'Fail'}")
```

```python
# Training parameters
learning_rate = 0.01  #step we take when updating weights (w) and bias (b) each iteration
epochs = 1000 #times we will go through the entire dataset to train the neuron.
```

```python
# Training loop
for epoch in range(epochs):
    total_loss = 0
    dw = np.zeros(2)
    db = 0
    
    for i in range(len(input)):
        x_i = input[i]
        y_i = output[i]
        
        # Forward pass
        y_hat = neuron_forward(x_i)
    
```

```python
 # Compute loss (binary cross-entropy) ->(when activation=sigmoid, output 1 or 0,two classes)
loss = - (y_i * np.log(y_hat + 1e-8) + (1 - y_i) * np.log(1 - y_hat + 1e-8))
total_loss += loss
```

```python
 # Compute loss (binary cross-entropy) ->(when activation=sigmoid, output 1 or 0,two classes)
loss = - (y_i * np.log(y_hat + 1e-8) + (1 - y_i) * np.log(1 - y_hat + 1e-8))
total_loss += loss
```

```python
 # Gradients
dz = y_hat - y_i
dw += dz * x_i
db += dz
```

```python
 # Update weights and bias
w -= learning_rate * dw / len(input)
b -= learning_rate * db / len(input)
    
if (epoch+1) % 100 == 0:
    print(f"Epoch {epoch+1}, Loss: {total_loss/len(input):.4f}")
```
